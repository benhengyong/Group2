```python
# Create dataframe contain the sub-industry for the food industry
for state in state_list:

    chosen_state = state

    chosen_state_sub_industry_df = data_by_state_and_industry.filter(regex= f'Month|{chosen_state}')

    chosen_state_sub_food_df = chosen_state_sub_industry_df.filter(regex= f'Month|Supermarket and grocery stores|Liquor retailing|Other specialised food retailing')
    
    # From chosen_state_sub_food_df, create a new dataframe contains only data from the last decade
    chosen_state_sub_food_df = chosen_state_sub_food_df.set_index('Month')

    # From chosen_state_df, create a new dataframe contains only data from the last decade
    chosen_state_sub_food_df = chosen_state_sub_food_df.loc['2013-01-01':'2023-09-01']

# Create a line chart showing the industry turnover by time for the chosen state in the last decade
    chosen_state_sub_food_df.plot(kind='line')

    plt.title(f"The sub-food industry turnover of {chosen_state} ")
    plt.xlabel('Time')
    plt.ylabel('Turnover ($ Millions)')
    plt.legend()
    plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")

    plt.show()
```
